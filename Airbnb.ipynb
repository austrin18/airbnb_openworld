```python
import pandas as pd
import numpy as np
```

```python
airbnb = pd.read_csv("abnb.csv", dtype={'id':str,
                                       'NAME':str,
                                       'host id':str,
                                       'host_identity_verfied':str,
                                       'host name': str,
                                       'neighbourhood group':str,
                                       'neighbourhood':str,
                                       'lat':float,
                                       'long':float,
                                       'country':str,
                                       'instant_bookable':str,
                                       'cancellation_policy':str,
                                       'room type':str,
                                       'Construction year':str,
                                       'price': int,
                                       'service fee': int,
                                       'minimum nights': int,
                                       'numberof reviews': int,
                                       'last review': str,
                                       'reviews per month': float,
                                       'review rate number': int,
                                       'calculated host listings count': int,
                                       'availability 365': int,
                                        'house_rules': str,
                                        'license': str,
                                       })
```

```python
df_airbnb = pd.DataFrame(airbnb)
print(df_airbnb)
```

```python
df_airbnb.head()
```

```python
df_airbnb.info()
```

```python
df_airbnb['id'] = df_airbnb['id'].astype(str)
df_airbnb['name'] = df_airbnb['name'].astype(str)
df_airbnb['host_id'] = df_airbnb['host_id'].astype(str)
df_airbnb['host_name'] = df_airbnb['host_name'].astype(str)
df_airbnb['neighbourhood_group'] = df_airbnb['neighbourhood_group'].astype(str)
df_airbnb['neighbourhood'] = df_airbnb['neighbourhood'].astype(str)
df_airbnb['room_type'] = df_airbnb['room_type'].astype(str)
df_airbnb['city'] = df_airbnb['city'].astype(str)
```

```python
df_airbnb.info()
```

```python
df_airbnb["last_review"] = pd.to_datetime(df_airbnb["last_review"])
```
